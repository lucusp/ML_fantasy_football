### Get Total Defense Data from ESPN ###

```python
d = []

for y in range(2010, 2021):
    alt_ = pd.read_html(f'https://www.espn.com/nfl/stats/team/_/view/defense/season/{y}/seasontype/2/', header=0)
    data = alt_[1].iloc[1:, :].reset_index().drop('index', axis=1)
    data.columns = alt_[1].loc[0]
    data = alt_[0].join(data)
    data['Year'] = y
    d.append(data)
    
def_data = pd.concat(d)

def_data.to_excel('def_data.xlsx', index=False)
```

### Get Total Offense Data from ESPN ###

```python
d = []

for y in range(2010, 2021):
    alt_ = pd.read_html(f'https://www.espn.com/nfl/stats/team/_/season/{y}/seasontype/2/', header=0)
    data = alt_[1].iloc[1:, :].reset_index().drop('index', axis=1)
    data.columns = alt_[1].loc[0]
    data = alt_[0].join(data)
    data['Year'] = y
    d.append(data)
    
off_data = pd.concat(d)

off_data.to_excel('off_data.xlsx', index=False)
```
